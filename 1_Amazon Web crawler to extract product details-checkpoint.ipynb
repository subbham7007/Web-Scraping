```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import time

# Step 1: Set up the web driver
driver = webdriver.Chrome()  # Replace with your ChromeDriver path

# Step 2: Function to parse and extract product details and store them in a list
def parse_and_extract(html):
    soup = BeautifulSoup(html, 'html.parser')
    results = soup.find_all('div', {'data-component-type': 's-search-result'})
    data = []

    for result in results:
        try:
            # Extract product name
            name = result.h2.text.strip()
        except:
            name = "N/A"
        
        try:
            # Extract seller (if available)
            seller = result.find('span', {'class': 'a-size-base-plus'}).text.strip()
        except:
            seller = "N/A"
        
        try:
            # Extract price (if available)
            price_whole = result.find("span", {"class": "a-price"}).find("span", {"class": "a-offscreen"}).text.strip()
            price = f"{price_whole}"
        except:
            price = "N/A"
        
        try:
            # Extract Prime status
            prime = result.find('i', {'aria-label': 'Amazon Prime'})
            prime_status = "Yes" if prime else "No"
        except:
            prime_status = "No"
        
        try:
            # Extract description (first few words of product name as a description)
            description = ' '.join(name.split(' ')[:5])
        except:
            description = "N/A"
        
        # Append extracted data as a list
        data.append([name, seller, price, prime_status, description])
    
    return data

# Step 3: Initialize variables
all_data = []
number_of_pages = 3
keyword = "shoe"
base_url = "https://amazon.in"

# Step 4: Create an empty DataFrame with the required columns
df = pd.DataFrame(columns=["Name", "Seller", "Price", "Prime", "Description"])

# Step 5: Loop through pages and scrape data
for page in range(1, number_of_pages + 1):
    print("page no =", page)
    try:
        if page == 1:
            search_url = f"https://www.amazon.in/s?k={keyword}&crid=238C8INH7USWD&qid=1733147112&sprefix=shoe%2Caps%2C458&ref=sr_pg_{page}"
        
        else:
            search_url = f"https://www.amazon.in/s?k={keyword}&page={page}&crid=238C8INH7USWD&qid=1732993188&sprefix=shoe%2Caps%2C458&ref=sr_pg_{page}"
        # "{base_url}/s?k={keyword}&page={page}&ref=nb_sb_noss"
        driver.get(search_url)
        
        # Step 6: Allow time for the page to load
        time.sleep(3)  # Increased sleep time to ensure full page load
        
        # Step 7: Parse the page content and get data as a list
        html = driver.page_source
        page_data = parse_and_extract(html)
        
        # Step 8: Convert the list to a DataFrame and append it to the main DataFrame
        page_df = pd.DataFrame(page_data, columns=["Name", "Seller", "Price", "Prime", "Description"])
        df = pd.concat([df, page_df], ignore_index=True)
        
    except Exception as e:
        print(f"Error on page {page}: {str(e)}")
        break

# Step 9: Print the DataFrame or save it to a CSV file
print(df.shape)
# df.to_csv("amazon_products.csv", index=False)

# Step 10: Close the driver
driver.quit()
```
